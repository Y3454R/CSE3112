```python
import numpy as np
import random

import matplotlib.pyplot as plt
```

```python
def f(x,y):
    #return ((x*x*x)-(10*x*x)+15)
    return x*x+y*y
```

```python
n_particles = 30
n_iter = 1000

#hyper parameters
w = 0.6
c1 = 2.1
c2 = 4-c1
```

```python
x = np.array([(-1) ** bool(random.getrandbits(1)) * random.random() * 5 for _ in range(n_particles)])
y = np.array([(-1) ** bool(random.getrandbits(1)) * random.random() * 5 for _ in range(n_particles)])

particle = np.array([x,y]).T
print(x)
print(particle)

v1 = np.array([(-1) ** bool(random.getrandbits(1)) * random.random() * 10 for _ in range(n_particles)])
v2 = np.array([(-1) ** bool(random.getrandbits(1)) * random.random() * 10 for _ in range(n_particles)])
velocity = np.array([v1,v2]).T

p_best_particle = particle.copy()
p_best = np.array([f(particle[i][0], particle[i][1]) for i in range(n_particles)])
g_best = float('inf')
g_x = particle[0]
```

```python
print(g_best,g_x)
```

```python
for iter in range (n_iter):
    #update local best 
    for i in range(n_particles):
        fit = f(particle[i][0],particle[i][1])
        if(p_best[i]>fit):
            p_best[i] = fit
            p_best_particle[i] = particle[i]
    
    #update global best
    for i in range(n_particles):
        fit = f(particle[i][0],particle[i][1])
        if(g_best>fit):
            g_best = fit
            g_x = particle[i]
    if(g_best<0.000000000000000001):
        print(f"Broken on iteration {iter} with value {g_best}")
        break
        
    #plot points
    if(iter%3==0):
        print(iter, 'iterations')
        print('BestPosition in this time:', g_x)
        print('BestValue in this time:', g_best)
        for i in range(n_particles):
            plt.plot(particle[i][0],particle[i][1],'ro')
        plt.plot(g_x[0],g_x[1], 'bo')
        plt.xlim(-5,5)
        plt.ylim(-5,5)
        plt.show()
        
        
    #update velocity
    new_v = []
    for i in range(n_particles):
        inertial = w * velocity[i]
        self_confidence = c1 * random.random() * (p_best_particle[i] - particle[i])
        swarm_confidence = c2 * random.random() * (g_x - particle[i])
        t_v = inertial + self_confidence + swarm_confidence
        if(t_v[0]>10):
            t_v[0] = 10.0
        elif(t_v[0]<-10):
            t_v[0] = -10.0
        if(t_v[1]>10):
            t_v[1] = 10.0
        elif(t_v[1]<-10):
            t_v[1] = -10.0
        new_v.append(t_v)
    velocity = np.array(new_v)
    particle = particle+velocity
        
```
